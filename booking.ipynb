Identify at least two variables that are likely decision-relevant and briefly justify your selection.

#### **Variable 1: Vehicle_Type**
* **Operational Justification:** This variable is essential for asset allocation and supply-demand balancing decisions within fleet operations. By evaluating the descriptive frequency and percentage distribution of ride requests across various vehicle categories (such as Prime Sedan, eBike, Auto, and Prime SUV), operations managers can identify which vehicle classes dominate consumer preferences and which choices are underutilized.
* **Business Impact:** If the descriptive analysis reveals a heavy structural demand for lower-cost options like Auto or eBike compared to premium tiers, executive management can adjust strategic decisions. This includes shifting driver recruitment campaigns, adjusting commission rate structures, or partnering with fleet manufacturers to expand high-demand vehicle categories to maximize total trip fulfillment rates.

#### **Variable 2: Payment_Method**
* **Operational Justification:** This variable directly informs financial operations, transaction cost management, and platform cash flow decisions. Descriptively tracking the frequency and percentage distribution of settlement channels (such as Cash, UPI, Credit Card, and Debit Card) allows financial managers to assess consumer payment preferences and identify platform processing vulnerabilities.
* **Business Impact:** A high reliance on physical cash payments implies operational friction, such as drivers needing physical change, increased risk of cash handling disputes, or delayed collection of the platform's commission fees. By analyzing these proportions, digital product managers can make data-driven decisions to launch targeted promotional discounts, partner with specific banking gateways, or optimize the application interface to incentivize digital checkouts (like UPI) and streamline cash collection cycles.
